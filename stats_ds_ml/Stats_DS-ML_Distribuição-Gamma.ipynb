# Distribuição Gamma

* Distribuição geral com valores assimétricos à direita
* Análise de tempo de vida de produtos
* Inadimplência com valores agregados
* Quantidade de chuva acumulada em um reservatório
* Tempo de reação de um motorista de acordo com a idade

```python
from scipy.stats import gamma
```

```python
data_gamma = gamma.rvs(a = 4, size = 1000)
sns.distplot(data_gamma);
```

```python
min(data_gamma), max(data_gamma)
```

```python
df = pd.read_csv('CSVs/credit_data.csv')
df.dropna(inplace = True)
df.shape
```

## Distribuição normal com o CSV 'credit_data.csv'

```python
df.head()
```

```python
X = df.iloc[:,:4].values
X
```

```python
y = df.iloc[:,4].values
y
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
result_nb = []
for i in range(30):
    X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2,
                                                    stratify=y, random_state=i)
    nb = GaussianNB()
    nb.fit(X_train, y_train)
    result_nb.append(accuracy_score(y_test, nb.predict(X_test)))
```

```python
print(result_nb)
```

```python
sns.distplot(result_nb, bins = 5);
```

# Variáveis discretas

## Distribuição Bernoulli

* Tipo específico da distribuição binomial
* Possui somente duas respostas: sucesso ou fracasso

```python
from scipy.stats import bernoulli
```

```python
data_bernoulli = bernoulli.rvs(size = 1000, p = 0.3)
data_bernoulli
```

```python
np.unique(data_bernoulli, return_counts = True)
```

```python
sns.distplot(data_bernoulli, kde = False);
```

## Distribuição binomial

* Probabilidade de sucesso ou falha

```python
from scipy.stats import binom
```

```python
data_bin = binom.rvs(size = 1000, n = 10, p = .8)
print(data_bin)
```

## Distribuição de Poisson

* Número de vezes que um evento aconteceu em um intervalo de tempo
* Exemplo: número de usuários que visitaram um website em um intervalo
* Quantidade de vezes que um evento aconteceu nesse intervalo (número de ligações em um call center)

```python
from scipy.stats import poisson
```

```python
data_pss = poisson.rvs(size = 1000, mu = 5)
```

```python
min(data_pss), max(data_pss)
```

```python
np.unique(data_pss, return_counts=True)
```

```python
sns.distplot(data_pss);
```
