O objetivo deste exercício é calcular o coeficiente e a taxa de evasão utilizando a tabela abaixo. Faça os cálculos manuais e depois implemente em Python



```python
print('TAXA DE EVASÃO (1/100)')
print(f'1º - {round(((70 - 65) / 70) * 100, 2)} para cada 100 alunos')
print(f'2º - {round(((50 - 48) / 50) * 100, 2)} para cada 100 alunos')
print(f'3º - {round(((47 - 40) / 47) * 100, 2)} para cada 100 alunos')
print(f'4º - {round(((23 - 22) / 23) * 100, 2)} para cada 100 alunos')
print(f'Total - {round(((190 - 175) / 190) * 100, 2)} para cada 100 alunos')
```

```python
import pandas as pd
```

```python
data = {'Ano_graduação':['1º','2º','3º','4º','Total'],
       'Mat_marco': [70,50,47,23,190],
       'Mat_novembro': [65,48,40,22,175]}
df = pd.DataFrame(data)
df.set_index('Ano_graduação', inplace=True)
df
```

```python
coef = (df.Mat_marco - df.Mat_novembro) / df.Mat_marco
coef
```

```python
taxa = coef * 100
df['%_taxa'] = taxa
df
```
