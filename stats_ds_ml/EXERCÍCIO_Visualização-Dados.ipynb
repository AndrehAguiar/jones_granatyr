```python
import os
import pandas as pd
import matplotlib.pyplot as plt 

os.environ["PROJ_LIB"] = '/share/proj'
from mpl_toolkits.basemap import Basemap
```

```python
df = pd.read_csv('CSVs/house_prices.csv')
df.head()
```

```python
df = df.sort_values(by='price', ascending=True)
df.head()
```

```python
df_baratos = df[:1000]
df_baratos
```

```python
df = df.sort_values(by='price', ascending=False)
df.head()
```

```python
df_caros = df[:1000]
df_caros
```

```python
lat1, lat2 = df['lat'].min(), df['lat'].max()
lon1, lon2 = df['long'].min(), df['long'].max()
```

```python
plt.figure(figsize=(10,10))
m = Basemap(projection ='cyl', resolution='h',
           llcrnrlat = lat1, urcrnrlat=lat2,
           llcrnrlon = lon1, urcrnrlon = lon2)

m.drawcoastlines()
m.fillcontinents(color='palegoldenrod', lake_color='lightskyblue')
m.drawmapboundary(fill_color ='lightskyblue')
m.scatter(df['long'], df['lat'], s = 5, c = 'green', alpha=0.1, zorder = 2)
m.scatter(df_caros['long'], df_caros['lat'], s = 10, c = 'red', alpha=0.1, zorder = 3)
m.scatter(df_baratos['long'], df_baratos['lat'], s = 10, c = 'blue', alpha=0.1, zorder = 4);
```
