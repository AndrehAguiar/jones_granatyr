Agora que você aprendeu o básico sobre correlação, é hora de aplicar esse conhecimento em uma base de dados real! Carregue a base de dados das casas (`house-prices.csv`) e gere as correlações entre todos os atributos. Quais atributos estão mais relacionados com o preço das casas?

```python
import numpy as np
import pandas as pd
import seaborn as sns
import math
```

```python
df = pd.read_csv('CSVs/house_prices.csv')
```

```python
df.head()
```

```python
df.drop(['id','date'], axis=1).corr() \
.style.background_gradient('Reds')
```

```python
sns.scatterplot(df['sqft_living'], df['price']);
```

O único atributo que a correlação com o "price" pode ser considerada forte é "sqft_living", as correlações consideradas moderadas são "bathrooms", "grade",	"sqft_above" e "sqft_living15". Todas as outras correlações com o atributo "price" são consideradas fracas e bem fracas.

```python
sns.scatterplot(df['bathrooms'], df['price']);
```

```python
sns.scatterplot(df['grade'], df['price']);
```

```python
sns.scatterplot(df['sqft_above'], df['price']);
```
